```python
import sympy as sp
from sympy import symbols, nonlinsolve
n,s,q,a = sp.symbols('n, s, q, a')
mu, rho, beta, sigma, gamma, T = sp.symbols('mu, rho, beta, sigma, gamma, T')
f1 = mu * (1 - rho) - beta * (s*n) - mu * n
f2 = beta * (s*n) - rho * (s*a) + sigma * (s*q*T) - mu * s
f3 = rho * (s*a) - sigma * (s*q*T) - gamma*q - mu*q
f4 = mu*rho + gamma*q - mu*a
solution=sp.nonlinsolve([f1,f2,f3,f4],[n,s,q,a])
print(solution)
#sp.solvers.solve((f1, f2, f3, f4), (n,s,q,a))
#https://stackoverflow.com/questions/68819409/solve-non-linear-system-of-equations-in-sympy
```
